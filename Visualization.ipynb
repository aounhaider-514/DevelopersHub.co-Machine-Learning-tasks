## **Exploring and Visualizing a Simple Dataset**

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style
sns.set(style='whitegrid')
%matplotlib inline
```

## Loading Dataset

```python
# REPLACE WITH YOUR ACTUAL FILE PATH
file_path = "C:/Users/Aoun Haider/Downloads/Iris.csv" 

# Load dataset
df = pd.read_csv(file_path)

# Verify load
print("Dataset loaded successfully!")
```

## Basic Dataset Inspection

```python
# Shape and columns
print("Dataset shape:", df.shape)
print("\nColumn names:")
print(df.columns.tolist())

# First 5 rows
print("\nFirst 5 rows:")
display(df.head())
```

## Summary Statistics

```python
# Dataset info
print("\nDataset Information:")
df.info()

# Descriptive statistics
print("\nDescriptive Statistics:")
display(df.describe())
```

## Scatter Plots

```python
# Scatter plot matrix
plt.figure(figsize=(10, 8))
sns.pairplot(df, hue='Species', vars=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'], 
             diag_kind=None, palette='viridis')
plt.suptitle('Feature Relationships', y=1.02)
plt.show()

# Individual scatter plot example
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='SepalLengthCm', y='PetalLengthCm', hue='Species', 
                palette='viridis', s=80)
plt.title('Sepal Length vs Petal Length')
plt.show()
```

##  Histograms

```python
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
plt.figure(figsize=(12, 8))

for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.histplot(data=df, x=feature, kde=True, hue='Species', 
                 element='step', palette='viridis', alpha=0.5)
    plt.title(f'{feature} Distribution')
    plt.xlabel(feature)
    
plt.tight_layout()
plt.show()
```

## Box Plots

```python
plt.figure(figsize=(12, 8))

for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=df, x='Species', y=feature, palette='viridis')
    plt.title(f'{feature} Distribution by Species')
    plt.xlabel('Species')
    plt.ylabel(feature)
    
plt.tight_layout()
plt.show()
```
